```python
 #first of all dont forget the use of data base
import mysql.connector as connection

try:
    mydb = connection.connect(host= "localhost",database = "amir789", user= "root",passwd="srk123",use_pure= True)
    #chech if the connection is established
    print(mydb.is_connected())
    
    query = "CREATE TABLE amirDetails (srkid int(10) AUTO_INCREMENT PRIMARY KEY,FirstName VARCHAR (60),"\
    "LastName VARCHAR(60),RegistrationDate DATE,Class int(12),Section Varchar(10))"
    
    cursor= mydb.cursor() #create a cursor to execute  queries
    cursor.execute(query)
    print("(table Created!!")
    mydb.close()
except Exception as e:
    mydb.close()
    print(str(e))
```

```python
import mysql.connector as connection

try:
    mydb = connection.connect(host= "localhost",database = "amir789", user= "root",passwd="srk123",use_pure= True)
    #chech if the connection is established
    print(mydb.is_connected())
    
    query = "insert into amirDetails VALUES('1000','amir','beg','1995-07-03','11', 'A')"
    
    cursor= mydb.cursor() #create a cursor to execute  queries
    cursor.execute(query)
    print("(values inserted into a table!!")
    mydb.close()
except Exception as e:
    mydb.close()
    mydb.commit()
    print(str(e))
    
```
